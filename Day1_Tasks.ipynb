```python

# Task 1: Numeric, Float, Complex, Boolean
int_positive = 10
int_negative = -5
int_binary = 0b1010
int_octal = 0o12
int_hexa = 0xA
float_num = 3.14
complex_num = 2 + 3j
bool_true = True
bool_false = False
print("Task 1 Outputs:")
print(int_positive, int_negative, int_binary, int_octal, int_hexa)
print(float_num)
print(complex_num)
print(bool_true, bool_false)
```

```python
# Task 2: Arithmetic Operators
a = 10
b = 3
print("Task 2 Outputs:")
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
print()
```
