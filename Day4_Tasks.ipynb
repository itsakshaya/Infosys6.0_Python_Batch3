```python
# Task1(a) Recursive function for Fibonacci
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        seq = fibonacci(n-1)
        seq.append(seq[-1] + seq[-2])
        return seq

num = int(input("Enter a number: "))
print("Fibonacci sequence:", fibonacci(num))
```

```python
# Task1(b) Recursive function for Factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a number to find factorial: "))
print(f"Factorial of {num} is:", factorial(num))
```

```python
# Task2- Write example for module OS, MATH, RANDOM
#Module Example

def greet(name):
    return f"Hello, {name}!"

def add(a, b):
    return a + b

print("Module Example:")
print(greet("Akshaya"))
print("5 + 3 =", add(5, 3))
print("\n")

# OS Module Example
import os
print("OS Module Example:")
print("Current Directory:", os.getcwd())
print("Files here:", os.listdir())
print("\n")


# Math Module Example
import math
print("Math Module Example:")
print("Square root of 16:", math.sqrt(16))
print("Factorial of 5:", math.factorial(5))
print("Pi value:", math.pi)
print("\n")


# Random Module Example
import random
print("Random Module Example:")
print("Random integer 1-10:", random.randint(1, 10))
colors = ["red", "blue", "green"]
print("Random color:", random.choice(colors))
```

```python
#Task3- list, tuple & dictionary comprehensions

# List Comprehension

squares_list = [x**2 for x in range(1, 6)]
print("List of squares:", squares_list)

even_squares_list = [x**2 for x in range(1, 6) if x % 2 == 0]
print("Even squares in list:", even_squares_list)
print("\n")


# Tuple Comprehension (using generator)
squares_tuple = tuple(x**2 for x in range(1, 6))
print("Tuple of squares:", squares_tuple)

even_squares_tuple = tuple(x**2 for x in range(1, 6) if x % 2 == 0)
print("Even squares in tuple:", even_squares_tuple)
print("\n")

# Dictionary Comprehension
square_dict = {x: x**2 for x in range(1, 6)}
print("Dictionary of squares:", square_dict)

even_square_dict = {x: x**2 for x in range(1, 6) if x % 2 == 0}
print("Even number squares in dictionary:", even_square_dict)
```
