```python
# Day 3- Task 1: Calculator with all 7 operators

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Select operator: +  -  *  /  %  //  **")
op = input("Enter operator: ")

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    print("Result:", a / b)
elif op == "%":
    print("Result:", a % b)
elif op == "//":
    print("Result:", a // b)
elif op == "**":
    print("Result:", a ** b)
else:
    print("Invalid operator")
```

```python
#Task 2: for loop using tuple, dictionary, set

numbers = (10, 20, 30, 40)
print("Tuple values:")
for n in numbers:
    print(n)

print("\nDictionary values:")
student = {"name": "Akshaya", "age": 21, "course": "MCA"}
for key in student:
    print(key, ":", student[key])

print("\nSet values:")
fruits = {"Python", "Java", "C++"}
for f in fruits:
    print(f)
```

```python
# Task 3: Calculator using structural pattern matching

def calculator(num1, num2, operator):
    match operator:
        case "+":
            return num1 + num2
        case "-":
            return num1 - num2
        case "*":
            return num1 * num2
        case "/":
            if num2 != 0:
                return num1 / num2
            else:
                return "Error: Division by zero!"
        case _:
            return "Invalid operator"

print("Simple Calculator")
n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

result = calculator(n1, n2, op)
print("Result:", result)
```

```python
# Task4- Simple File Handling Demo

# Write to a file
file = open("demo.txt", "w")
file.write("Hello World!\nPython is fun!\n")
file.close()

# Append to the file
file = open("demo.txt", "a")
file.write("This line is added.\n")
file.close()

# Read the entire file
file = open("demo.txt", "r")
print("Full content:\n", file.read())
file.close()

# Read line by line
file = open("demo.txt", "r")
print("\nLine by line:")
for line in file:
    print(line.strip())
file.close()

# Rename the file
import os
os.rename("demo.txt", "demo_renamed.txt")

# Delete the file
os.remove("demo_renamed.txt")
print("\nFile renamed and then deleted successfully!")
```
