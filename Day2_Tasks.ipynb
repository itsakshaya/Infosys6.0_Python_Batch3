```python
# Task 1: Google Form with all data types
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height: "))
num = complex(input("Enter a complex number: "))
is_student = input("Are you a student? True/False: ")

# Method 1: Comma
print("Using comma:", name, age, height, num, is_student)

# Method 2: Plus (+)
print("Using plus: " + name + " " + str(age) + " " + str(height) + " " + str(num) + " " + str(is_student))

# Method 3: Percentage (%)
print("Using %%: Name=%s Age=%d Height=%.2f Complex=%s Student=%s" % (name, age, height, num, is_student))

# Method 4: format()
print("Using format: Name={} Age={} Height={} Complex={} Student={}".format(name, age, height, num, is_student))

# Method 5: f-string
print(f"Using f-string: Name={name} Age={age} Height={height} Complex={num} Student={is_student}")
```

```python
# Task 2: Functions manipulating lists

my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)

# 1. append()
my_list.append(60)
print("After append:", my_list)

# 2. insert()
my_list.insert(2, 15)
print("After insert:", my_list)

# 3. remove()
my_list.remove(40)
print("After remove:", my_list)

# 4. pop()
my_list.pop(3)
print("After pop:", my_list)

# 5. sort()
my_list.sort()
print("After sort:", my_list)

# 6. reverse()
my_list.reverse()
print("After reverse:", my_list)

# 7. len()
print("Length of list:", len(my_list))

# 8. index()
print("Index of 50:", my_list.index(50))

# 9. count()
print("Count of 20:", my_list.count(20))

# 10. clear()
my_list.clear()
print("After clear:", my_list)
```

```python
# Task 3: Built-in functions for Dictionary

my_dict = {"name": "Jyothi", "age": 21, "course": "MCA"}
print("Original Dictionary:", my_dict)

# 1. keys()
print("Keys:", my_dict.keys())

# 2. values()
print("Values:", my_dict.values())

# 3. items()
print("Items:", my_dict.items())

# 4. get()
print("Get age:", my_dict.get("age"))

# 5. update()
my_dict.update({"college": "Aditya University"})
print("After update:", my_dict)

# 6. pop()
my_dict.pop("age")
print("After pop:", my_dict)

# 7. popitem()
my_dict.popitem()
print("After popitem:", my_dict)

# 8. len()
print("Length:", len(my_dict))

# 9. clear()
my_dict.clear()
print("After clear:", my_dict)
```

```python
# Task 4: Functions to modify set

# Original set
my_set = {10, 20, 30, 40, 50}
print("Original Set:", my_set)

# 1. add()
my_set.add(60)
print("After add:", my_set)

# 2. update()
my_set.update([70, 80])
print("After update:", my_set)

# 3. remove()
my_set.remove(20)
print("After remove:", my_set)

# 4. discard()
my_set.discard(100)
print("After discard:", my_set)

# 5. pop()
removed_element = my_set.pop()
print("After pop:", my_set, "| Removed:", removed_element)

# 6. clear()
my_set.clear()
print("After clear:", my_set)
```

```python
# Task 5: Set Operations

set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

print("Set 1:", set1)
print("Set 2:", set2)

# 1. union()
print("Union:", set1.union(set2))

# 2. intersection()
print("Intersection:", set1.intersection(set2))

# 3. difference()
print("Difference (set1 - set2):", set1.difference(set2))

# 4. symmetric_difference()
print("Symmetric Difference:", set1.symmetric_difference(set2))

# 5. issubset()
print("Is set1 subset of set2?:", set1.issubset(set2))
```
